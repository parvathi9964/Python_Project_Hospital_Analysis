```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/parva/OneDrive/Desktop/COACHX LIVE/PYTHON/PYTHON_PRACTISE/Hospital_Anlaysis/Healthcare_Operations_Dataset.csv")
df.head()
```

```python
df.describe()
df.info()
```

```python
dept = df['Department'].value_counts()

plt.bar(dept.index, dept.values)
plt.title("Patients by Department")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(df['TreatmentCost'], bins=30)
plt.title("Treatment Cost Distribution")
plt.show()
```

```python
sns.histplot(df["Age"], bins=30)
plt.title("Patient Age Distribution")
plt.show()
```

```python
sns.countplot(data=df,x="Department")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=df,x="Department",y="TreatmentCost")
plt.xticks(rotation=45)
plt.show()
```
